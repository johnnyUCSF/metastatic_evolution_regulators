# Convert encode_tf annotation to ensembl 

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
import scipy.stats
```

### import dictionary file and convert to dict
##### note that this is derived from hg19 ensembl biomart

```python
ensembl_dict = pd.read_csv('ensembl_encode_dict.txt',sep='\t',index_col=1)
ensembl_dict = ensembl_dict.to_dict()
ensembl_dict = ensembl_dict.get('Gene stable ID')
ensembl_dict
```

### convert annotation file from gene name to ensembl

```python
import re
########merge files together
with open('human_encode_tf_index.txt','r') as readfile, open('ensembl_human_encode_tf_index.txt','w') as out:
    for line in readfile:
        ####this line is useful for 
        line = re.sub("\s+$", "", line)
        arr = line.split("\t")
        gene = arr.pop(0)
        if gene in ensembl_dict:
            out.write(ensembl_dict[gene])
            out.write('\t')
            out.write('\t'.join(arr))
            out.write('\n')
```
