```python
############this is for generating bed file +/- 1kb of TSS
#####################################
biomart = pd.read_csv('ensembl_biomart_chrm_TSS.txt',sep='\t')
############add score
biomart['score']=1000
#####read into bedfile
bedfile = biomart[['Chromosome/scaffold name','Transcription start site (TSS)','Transcription start site (TSS)','Gene stable ID','score','Strand']]
bedfile.columns = ['chrom','chromStart','chromEnd','name','score','strand']
######check to see if chromosomes are standard (23 or X Y)
def check_chrom(chromosome):
    ####create list of chrom numbers
    tmp = range(1,23)
    chrom_list = []
    ####convert to string
    for item in tmp:
        chrom_list.append(str(item))
    ####
    chrom_list.append('X')
    chrom_list.append('Y')
    ####
    if chromosome in chrom_list:
        return True
    else:
        return False
######filter those that have proper chromosome
bedfile['keep'] = bedfile.apply(lambda row: check_chrom(row['chrom']),axis=1)
bedfile = bedfile[bedfile['keep']==True]
######
bedfile = bedfile.drop(['keep'],axis=1)
#####add 1kb plus and 1kb minus
bedfile['chrom'] = 'chr'+bedfile['chrom']
bedfile['chromStart'] = bedfile['chromStart']-1000
bedfile['chromEnd'] = bedfile['chromEnd']+1000
######save file
bedfile.to_csv('ensembl_TSS_2kb.bed',sep='\t',index=None,header=None)
#####
bedfile.head(n=1000)
```
